```python
df = yf.download(stocks, start=start, end=end)
```

```python
df = df["Adj Close"]
```

```python
# Get daily returns
daily_returns = df.pct_change()
```

```python
start = dt.date(2000,1,1)
end = dt.datetime.now()
stocks = ['AAPL', 'MSFT']

df = yf.download(stocks, start=start, end=end)
```

```python
df.head()
```

```python
df = df["Adj Close"]
df.head()
```

```python
df.plot()
```

```python
daily_returns = df.pct_change()
daily_returns
```

```python
daily_returns.plot()
```

```python
daily_cum_returns = (daily_returns+1).cumprod()
```

```python
daily_cum_returns
```

```python
daily_cum_returns.plot()
```
